```python
class TreeNode():
    def __init__(self, data):
        self.data = data
        self.children = []
        self.parent = None
    
    def add_child(self, child):
        child.parent = self
        self.children.append(child)
    
    def get_level(self):
        level = 0
        
        p = self.parent
        while p:
            level += 1
            p = p.parent
        
        return level
    
    def print_tree(self):
        lvl = self.get_level()
        spaces = " "*lvl
        branch = "|--" if lvl else ""
        print(spaces*2 + branch + self.data)
        if self.children:
            for child in self.children:
                child.print_tree()
        
    
def build_tree():
    
    laptop = TreeNode("Laptop")
    lenovo = TreeNode("Lenovo")
    laptop.add_child(lenovo)
    mac_book = TreeNode('Mac Book')
    laptop.add_child(mac_book)
    dell = TreeNode("Dell")
    laptop.add_child(dell)
    
    mobile = TreeNode("Mobile")
    iphone = TreeNode('iPhone')
    mobile.add_child(iphone)
    honor = TreeNode('Honor-7X')
    mobile.add_child(honor)
    one_plus = TreeNode("One Plus")
    mobile.add_child(one_plus)
    
    tv = TreeNode('TV')
    samsung = TreeNode('Samsung')
    tv.add_child(samsung)
    mi = TreeNode("MI TV")
    tv.add_child(mi)
    
    root = TreeNode("Electronics")
    root.add_child(laptop)
    root.add_child(mobile)
    root.add_child(tv)
    
    return root

if __name__ == "__main__":
    root = build_tree()
    root.print_tree()
```
